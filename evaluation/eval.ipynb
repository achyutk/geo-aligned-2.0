```python
import torch
import clip
from PIL import Image
import os
import pycountry
import pandas as pd
import numpy as np
import utils
from collections import Counter
```

```python
path='C:/Users/ak19g21/Downloads/Project/output/'
folder_name = 'no_translation_wiki_w2v_nitro_sd'    #Change the name of the framework that you want to evaluate

files = [f for f in os.listdir(path+'/'+folder_name)][:1000]

translations = []
f = open(path+folder_name+"/translations.txt", 'r')
for x in f.readlines():
    translations.append(x)
f.close()

aligned_texts=[]
g = open(path+folder_name+"/aligned_texts.txt", 'r')
for x in g.readlines():
    aligned_texts.append(x)
g.close()
```

```python
model, preprocess = clip.load('ViT-B/32')
```

# Metric 1

```python
total_score = 0
for i in range(len(files)):
    image_path = path + folder_name + '/'+files[i]
    text = aligned_texts[i]
    score = utils.get_clip_score_img_text(model, preprocess,image_path, text)
    print(files[i],':',score)
    total_score = total_score + score
average_score = total_score/len(files)

average = ['Average_score:',average_score]
with open(path + folder_name + '/metric1.txt', 'w',encoding="utf8") as file:
    for item in average:
        file.write(str(item) + '\n')  
```

# Metric 2

```python
total_score = 0
for i in range(len(files)):
    image_path = path + folder_name + '/'+files[i]
    text = 'india'
    score = utils.get_clip_score_img_text(model, preprocess,image_path, text)
    print(files[i],':',score)
    total_score = total_score + score
average_score = total_score/len(files)

average = ['Average_score:',average_score]
with open(path + folder_name + '/metric2.txt', 'w',encoding="utf8") as file:
    for item in average:
        file.write(str(item) + '\n')  
```

# Metric 3

```python
text_feature_list = torch.load('model/eval_country_text_features.pt')
country =[]
f = open("model/eval_country_text_country_list.txt", 'r')
for x in f.readlines():
    country.append(x)
f.close()

country_prediction =[] 
for i in range(len(files)):
    image_path = path + folder_name + '/'+files[i]
    score = utils.get_clip_score_img_country(model,preprocess,image_path,text_feature_list,country)
    country_prediction.append(score[0])
    
counts = Counter(country_prediction)

sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)
with open(path + folder_name + '/metric3.txt', 'w') as f:
    for key, value in sorted_counts:
        key = key.rstrip('\n')
        f.write(f"{key} {value}\n")
```

# Metric 4

```python
img2_feature_list = torch.load('model/eval_img_features.pt')
country = []
with open('model/eval_img_features_country_list.txt', 'r') as file:
    for line in file:
        country.append(line.strip())
        
country_prediction =[] 
for i in range(len(files)):
    image_path = path + folder_name + '/'+files[i]
    score = utils.get_similarity_score_img_img(model, preprocess,image_path,country,img2_feature_list)
    country_prediction.append(score[0])
    
counts = Counter(country_prediction)

sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)
with open(path + folder_name + '/metric4.txt', 'w') as f:
    for key, value in sorted_counts:
        key = key.rstrip('\n')
        f.write(f"{key} {value}\n")
```
